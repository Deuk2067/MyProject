```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False

import pandas as pd
```

```python
K_best = pd.read_csv('./data/K_Best.txt', index_col=0, parse_dates=['최신 연재일', '연재 시작일'])
K_nor = pd.read_csv('./data/K_Normal.txt', index_col=0, parse_dates=['최신 연재일', '연재 시작일'])
M_best = pd.read_csv('./data/M_Best.txt', index_col=0, parse_dates=['마지막 연재일', '연재 시작일'])
M_nor = pd.read_csv('./data/M_Normal.txt', index_col=0, parse_dates=['마지막 연재일', '연재 시작일'])
N_best = pd.read_csv('./data/N_Best.txt', index_col=0, parse_dates=['마지막 연재일', '연재 시작일'])
N_nor = pd.read_csv('./data/N_Normal.txt', index_col=0, parse_dates=['마지막 연재일', '연재 시작일'])
```

```python
K_best.rename(columns={'최신 연재일': '마지막 연재일'}, inplace=True)
K_best.info()
```

```python
K_nor.rename(columns={'최신 연재일': '마지막 연재일'}, inplace=True)
K_nor
```

```python
M_best=M_best.drop(columns=['현대판타지', '판타지', '무협','로맨스판타지'])
M_best.rename(columns={'회수': '연재수'}, inplace=True)
M_nor=M_nor.drop(columns=['현대판타지', '판타지', '무협','로맨스판타지','로맨스'])
M_nor.rename(columns={'회수': '연재수'}, inplace=True)
N_best=N_best.drop(columns=['현판', '판타지', '무협','로판','로맨스'])
N_best.rename(columns={'회수': '연재수'}, inplace=True)
N_nor.rename(columns={'회수': '연재수'}, inplace=True)
```

```python
Kdf = pd.concat([K_nor, K_best], axis=0, ignore_index=1)
Kdf.index = Kdf.index+1
Kdf
```

```python
Kdf=Kdf.drop_duplicates('제목', keep='first', ignore_index=True)
Kdf['Site'] = 'Kakao_Page'
Kdf
```

```python
Mdf = pd.concat([M_nor, M_best], axis=0, ignore_index=1)
Mdf.index = Mdf.index+1
Mdf
```

```python
Mdf=Mdf.drop_duplicates('제목', keep='first', ignore_index=True)
Mdf['Site'] = 'Munpia'
Mdf
```

```python
Ndf = pd.concat([N_nor, N_best], axis=0, ignore_index=1)
Ndf.index = Ndf.index+1
Ndf
```

```python
Ndf=Ndf.drop_duplicates('제목', keep='first', ignore_index=True)
Ndf['Site'] = 'NaverSeries'
Ndf.info()
```

```python
Mdf.info()
```

```python
Mdf.to_csv('./data/M_all.txt')
Ndf.to_csv('./data/N_all.txt')
Kdf.to_csv('./data/K_all.txt')
```

```python
Mdf = Mdf[['제목', '장르', '작가', '연재수', '연재 시작일', '마지막 연재일', 'Site']]
Ndf = Ndf[['제목', '장르', '작가', '연재수', '연재 시작일', '마지막 연재일', 'Site']]
Kdf = Kdf[['제목', '장르', '작가', '연재수', '연재 시작일', '마지막 연재일', 'Site']]

MNK_all = pd.concat([Mdf, Ndf, Kdf], axis=0, ignore_index=1)
MNK_all
```

```python
MNK_all=MNK_all.drop_duplicates('제목', keep='first', ignore_index=True)
MNK_all.장르.value_counts()
```

```python
MNK_all.loc[MNK_all.장르 == '현판', '장르'] = '현대판타지'
MNK_all.loc[MNK_all.장르 == '로판', '장르'] = '로맨스판타지'
MNK_all.장르.value_counts()
```

```python
MNK_all.to_csv('./data/All_Novel.txt')
```
