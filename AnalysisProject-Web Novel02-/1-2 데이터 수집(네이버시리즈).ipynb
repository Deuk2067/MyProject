```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

import numpy as np
import pandas as pd
import time

```

```python
targetSite = 'https://series.naver.com/novel/top100List.series' 
driver = webdriver.Chrome() 
driver.get(targetSite)
time.sleep(0.5)

for i in range (1,5,1): # 실시간, 일간, 주간, 월간
    driver.find_element(By.XPATH, f'//*[@id="container"]/div[1]/ul/li[{i}]/a').click()
    time.sleep(1)
    
    Titles = [] # 제목 저장할 리스트
    Genre = [] # 장르 저장할 리스트
    Author = [] # 작가 저장할 리스트
    Start = [] # 연재 시작일
    Last = [] # 마지막 연재일
    Count = [] # 연재수
    Good = [] # 추천수
    Point = [] # 평점
    Comment = [] # 댓글 수
    
    for p in range(1,6,1):
        if p == 1: # 첫 페이지 1~20위
            pass
        else: # p가 2부터5까지
            driver.find_element(By.XPATH, f'//*[@id="content"]/div/div[2]/a[{p-1}]').click() # 2, 3, 4, 5
            
        for j in range(1, 21, 1):
            # 1~20위, 21~40, 41~60 소설 클릭
            driver.find_element(By.XPATH, f'//*[@id="content"]/div/ul/li[{j}]/div[2]/h3/a').click() # j번째 소설 들어감

            if "login" in driver.current_url or "adult" in driver.current_url: # 19금, 로그인 필요한 경우
                # 이후 전처리 할 떄 쉽게 하도록 19금 통일
                Titles.append("19금")
                Genre.append("19금")
                Author.append("19금")            
                Point.append("19금")
                Start.append("19금")
                Last.append("19금")
                Count.append("19금")
                Good.append("19금")
                Comment.append("19금")
                driver.back()
                continue
        
            # 정렬 기준이 랜덤이라 Start, Last에 들어가는게 반대로 들어가는 경우가 있음 (최신순, 1화부터) 
            a=driver.find_element(By.XPATH, '//*[@id="content"]/table/thead/tr/th[1]/div') # 최신순
            b=driver.find_element(By.XPATH, '//*[@id="content"]/table/thead/tr/th[2]/div') # 1화부터

            if a.get_attribute('class') == 'on _changeTicketSortOrder(DESC)': # 내림차순 정렬, 최신순 정렬되있는 경우
                time.sleep(0.5)
                Last.append(driver.find_element(By.XPATH, '//*[@id="volumeList"]/tr[1]/td[1]/em').text) # 마지막 연재일  
                time.sleep(0.5)
                b.click() # 1화부터 체크 => 올림차순 정렬 1화부터 정렬
                time.sleep(0.5)
                Start.append(driver.find_element(By.XPATH, '//*[@id="volumeList"]/tr[1]/td[1]/em').text) # 연재 시작일 추가
                driver.back() # 뒤로 돌아감 = 다시 내림차순 정렬, 최신순 정렬로 돌아감

            elif b.get_attribute('class') == 'on _changeTicketSortOrder(ASC)': # 올림차순 정렬, 1화부터 정렬되 있을 경우
                time.sleep(0.5)
                Start.append(driver.find_element(By.XPATH, '//*[@id="volumeList"]/tr[1]/td[1]/em').text) # 연재 시작일 추가
                time.sleep(0.5)
                a.click() # 최신순부터 체크 => 내림차순 정렬 최신화부터 정렬
                time.sleep(0.5)
                Last.append(driver.find_element(By.XPATH, '//*[@id="volumeList"]/tr[1]/td[1]/em').text) # 마지막 연재일  
                driver.back() # 뒤로 돌아감 = 다시 올림차순 정렬, 1화부터 정렬
            
            try:
                Titles.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[1]/h2').text) # 제목
                Good.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[1]/div[2]/ul/li[2]/div/a/em').text) # 추천수
            except:
                Titles.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/h2').text) # 제목
                Good.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[2]/ul/li[2]/div/a/em').text) # 추천수
            try:
                Point.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[1]/div[1]/em').text) # 평점
            except:
                Point.append(driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/em').text) # 평점
            Genre.append(driver.find_element(By.XPATH, '//*[@id="content"]/ul[1]/li/ul/li[2]/span/a').text) # 장르
            Author.append(driver.find_element(By.XPATH, '//*[@id="content"]/ul[1]/li/ul/li[3]/a').text) # 작가
            Count.append(driver.find_element(By.XPATH, '//*[@id="content"]/h5/strong').text) # 연재수

            time.sleep(1)
            Comment.append(driver.find_element(By.XPATH, '//*[@id="cbox_module"]/div/div[1]/span').text)
            #print(len(Titles), len(Genre), len(Author), len(Start), len(Last), len(Count), len(Comment), len(Good), len(Point))
            print(f'p:{p}. {j}. 제목: {Titles[j-1]}, 장르: {Genre[j-1]}, 작가: {Author[j-1]}, 시작일: {Start[j-1]}, 최신화: {Last[j-1]}, 연재수: {Count[j-1]}, 추천수: {Good[j-1]}, 평점: {Point[j-1]}, 댓글: {Comment[j-1]}')
            driver.back()# i번 소설 크로링 후 전 페이지로    
    
    df= pd.DataFrame({"제목":Titles, "장르":Genre, "작가":Author, "연재 시작일":Start, "마지막 연재일":Last, "회수":Count, "댓글수":Comment, "추천수":Good, "평점": Point})
    df.index = df.index+1
    
    if i == 1:
        df.to_csv('./data/NSeries_Now.txt')
    if i == 2:
        df.to_csv('./data/NSeries_Today.txt')
    if i == 3:
        df.to_csv('./data/NSeries_Week.txt')
    else:
        df.to_csv('./data/NSeries_Month.txt')        
        
```
