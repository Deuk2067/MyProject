```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False

import pandas as pd
```

```python
df = pd.read_csv('./data/Kakao_최신.txt', index_col=0)
df
```

```python
# 19금 제거
index=df.loc[df.제목 == '19금'].index
df.drop(index, inplace=True)
df
```

```python
patterns = [
    r'\[단행본\]', r'\[독점\]', r'\[연재\]', r'\[선공개\]', r'\[오디오드라마\]',
    r'\[개정판\-단독선연재\]', r'\[개정판\]', r'\[무적행 개정판\]', r'\[완결\]', r'\[단독 선공개\]',
    r'\[단독선공개\]', r'\[|\]'
]

for pattern in patterns:
    df['제목'] = df['제목'].str.replace(pattern, '', regex=True).str.strip()
df    
```

```python
# 제목이 중복되는 자료 제거
df=df.drop_duplicates('제목', keep='first', ignore_index=True)
df.제목.value_counts() # 중복값 확인
```

```python
# '전체' 문자열 제거 및 공백 제거
df['댓글'] = df['댓글'].str.replace('전체', '', regex=True).str.strip()

# '만'과 '억' 단위 처리
df['댓글'] = df['댓글'].str.replace(r'(\d+\.\d+)만', lambda x: x.group(1).replace('.', '') + '000', regex=True)
df['댓글'] = df['댓글'].str.replace(r'(\d+)만', lambda x: x.group(1) + '0000', regex=True)
df['댓글'] = df['댓글'].str.replace(r'(\d+\.\d+)억', lambda x: x.group(1).replace('.', '') + '0000000', regex=True)
df['댓글'] = df['댓글'].str.replace(r'(\d+)억', lambda x: x.group(1) + '00000000', regex=True)

# 공백 및 불필요한 문자 제거
df['댓글'] = df['댓글'].str.replace('.', '', regex=True).str.strip()
df['댓글'] = df['댓글'].str.replace(',', '', regex=True).str.strip()

# 연재수 처리
df['연재수'] = df['연재수'].str.replace('전체', '', regex=True).str.strip()
df['연재수'] = df['연재수'].str.replace(',', '', regex=True).str.strip()

# 조회수 처리
df['조회수'] = df['조회수'].str.replace('전체', '', regex=True).str.strip()
df['조회수'] = df['조회수'].str.replace(r'(\d+\.\d+)만', lambda x: x.group(1).replace('.', '') + '000', regex=True)
df['조회수'] = df['조회수'].str.replace(r'(\d+)만', lambda x: x.group(1) + '0000', regex=True)
df['조회수'] = df['조회수'].str.replace(r'(\d+\.\d+)억', lambda x: x.group(1).replace('.', '') + '0000000', regex=True)
df['조회수'] = df['조회수'].str.replace(r'(\d+)억', lambda x: x.group(1) + '00000000', regex=True)
df['조회수'] = df['조회수'].str.replace(',', '', regex=True).str.strip()

# 연재 시작일 및 최신 연재일 처리
df['연재 시작일'] = df['연재 시작일'].str.replace(r'[()]', '', regex=True).str.strip()
df['최신 연재일'] = df['최신 연재일'].str.replace(r'[()]', '', regex=True).str.strip()

```

```python
df
```

```python
df.조회수 = df.조회수.astype(int)
df.연재수 = df.연재수.astype(int)
df.댓글 = df.댓글.astype(int)
df.평점 = df.평점.astype(float)
df['연재 시작일'] = pd.to_datetime(df['연재 시작일'])
df['최신 연재일'] = pd.to_datetime(df['최신 연재일'])
```

```python
df.info()
df
```

```python
df['장르'] = df['장르'].str.replace('웹소설', '', regex=True).str.strip()
```

```python
print(df['장르'].value_counts())

# 드라마 항목의 인덱스
index_num = df[df.장르 == '드라마'].index

# 장르와 개수
genres = ['현판', '판타지', '무협', '로판', '로맨스']
counts = [7, 8, 8, 8, 8]

# 드라마 항목 분배 # '현대판타지',0:39 , '판타지', 39:78 ~ 
start_idx = 0
for genre, count in zip(genres, counts):
    end_idx = start_idx + count
    df.loc[index_num[start_idx:end_idx], '장르'] = genre
    start_idx = end_idx
```

```python
df.loc[df.장르 == 'BL', '장르'] = '로맨스'
df.loc[df.장르 == '현판', '장르'] = '현대판타지'
df.loc[df.장르 == '로판', '장르'] = '로맨스판타지'
```
